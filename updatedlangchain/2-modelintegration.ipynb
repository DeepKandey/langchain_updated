```python
import os
from dotenv import load_dotenv
load_dotenv()

os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")    
```

```python
from langchain.chat_models import init_chat_model

model_openAI = init_chat_model(model="gpt-5-mini")
    
model_openAI.invoke("Hello, how are you?") 
```

```python
model_google = init_chat_model(model="google_genai:gemini-2.5-flash-lite")
    
model_google.invoke("Hello, how are you?") 
```

```python
model_groq = init_chat_model(model="groq:llama-3.1-8b-instant")
    
model_groq.invoke("How should testers deal with AI landscapes?") 
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model_gemini = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
model_gemini.invoke("What is nested learning?")
```

```python
# streaming
for chunk in model_groq.stream("Write me a 200 words on Deep Learning"):
    print(chunk.text,end="|",flush=True)
```

```python
# batching
responses = model_groq.batch([
    "Deep Learning in 100 words", 
    "Machine Learning in 100 words", 
    "Artificial Intelligence in 100 words"],
    config={
        "max_concurrency": 2
    })
for response in responses:
    # print(response.text)
    print(response)
```
